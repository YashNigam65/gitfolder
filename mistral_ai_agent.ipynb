```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

MODEL_NAME = "mistralai/Mistral-7B-Instruct-v0.2"  # change if you like

# Load tokenizer & model
print("Loading model... (first time may take a while)")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,   # use float32 if you don't have GPU
    device_map="auto",           # puts model on GPU if available
)

SYSTEM_PROMPT = (
    "You are a helpful AI agent. "
    "Answer clearly and concisely. "
    "Ask for clarification only when truly needed."
)

def generate_reply(user_input: str) -> str:
    # Simple chat-style formatting
    prompt = f"<s>[INST] <<SYS>>\n{SYSTEM_PROMPT}\n<</SYS>>\n{user_input} [/INST]"

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
        )

    full_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # For simple demos, just return everything after the user message
    # (In practice, you might parse more carefully.)
    return full_text.split(user_input)[-1].strip()

def run_agent():
    print("Simple open-source AI agent. Type 'quit' to exit.\n")

    while True:
        user_input = input("You: ")
        if user_input.strip().lower() in {"quit", "exit"}:
            print("Agent: Goodbye! 👋")
            break

        reply = generate_reply(user_input)
        print(f"Agent: {reply}\n")
```
