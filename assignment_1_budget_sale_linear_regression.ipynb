```python
step 1: import the libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

step 2: creating dataframe and input output

```python
data = {
    'TV_Budget': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7, 57.5, 120.2, 144.1, 111.6],
    'Radio_Budget': [37.8, 39.3, 45.9, 41.3, 10.8, 48.9, 32.8, 19.6, 16.0, 12.6],
    'Newspaper_Budget': [69.2, 45.1, 69.3, 58.5, 58.4, 75.0, 23.5, 11.6, 40.3, 37.9],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2, 11.8, 13.2, 15.6, 12.2]
}

df = pd.DataFrame(data)
display(df)
```

```python
inputx = df.iloc[:, :-1].values
print(inputx)

outputy = df.iloc[:, -1:].values
print(outputy)
```

step 3: The code uses a test size of 1/4, which means it splits the data into 75% for training and 25% for testing.

```python
input_train, input_test, output_train, output_test = train_test_split(inputx, outputy, test_size = 1/4, random_state = 7)
```

step 4: selecting the Linear Regression model and training your linear regression model.

```python
model = LinearRegression()

model.fit(input_train, output_train)
```

step 5: testing or model prediction using testinput

```python
TV_Budget = float(input("\nGive TV_Budget  "))
Radio_Budget = float(input("\nGive Radio_Budget  "))
Newspaper_Budget = float(input("\nGive Newspaper_Budget  "))

testinput = [[TV_Budget,Radio_Budget,Newspaper_Budget]]
print(testinput)
predicted_output = model.predict(testinput)
print('\nBudget is ',testinput)
print('\nproduct sales based on the advertising budget allocated to different channels ',predicted_output)
```

step 6: Visualising the training results

```python


# Visualizing the predicted sales vs actual sales for the training set
plt.figure(figsize=(6, 6))
plt.scatter(output_train, model.predict(input_train), color = 'green')
plt.plot([output_train.min(), output_train.max()], [output_train.min(), output_train.max()], 'k--', lw=2)
plt.title('Actual Sales vs Predicted Sales (Training set)')
plt.xlabel('Actual Sales (units)')
plt.ylabel('Predicted Sales (units)')
plt.show()
```
