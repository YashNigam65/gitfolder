This code cell attempts to install the latest available nightly builds of torch, torchvision, and torchaudio that are compatible with CUDA 12.1.
```python
!pip install gtts
from gtts import gTTS
import os
# Inspect the TTS object to find available methods
mytext = 'Welcome to Google Text-to-Speech, I am yash, How are you guys ?'
language = 'en'
myobj = gTTS(text=mytext, lang=language, slow=False)
myobj.save("welcome_one.mp3")
os.system("mpg321 welcome.mp3")
```
