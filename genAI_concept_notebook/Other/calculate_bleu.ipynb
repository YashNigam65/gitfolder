```python
!pip install sacrebleu
import sacrebleu

# 1. Reference (gold) sentences
references = [
    "The cat is on the mat.",
    "I like to eat apples.",
    "This is a BLEU score example."
]

# 2. Model (candidate) outputs
candidates = [
    "The cat is on the mat.",
    "I enjoy eating apples.",
    "This is an example of BLEU score."
]

# sacrebleu expects:
# - references: list of reference *lists* (one list per ref set)
# - candidates: list of hypothesis strings
bleu = sacrebleu.corpus_bleu(candidates, [references])

print(f"BLEU score: {bleu.score:.2f}")
```

BLEU Ranges:
"<10": bad,
10-20: weak
20-30: ok
30-40: good
40-50: very strong, often close to fluent human translations
 ">50": near human or sometime even higher then human
